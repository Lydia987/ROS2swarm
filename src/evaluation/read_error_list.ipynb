```python
import matplotlib.pyplot as plt

def plot_list(name):
    # ACHTUNG: Pfad noch anpassen
    with open('/home/lydia/error_list' + name + '.txt', 'r') as doc:
        error_list = eval(doc.read())

    times = []
    errors = []

    for point in error_list:
        times.append(point[0]-error_list[-1][0])
        errors.append(point[1])

    start = 200
    intervall = 250
    end = start + intervall # len(error_list)  # 233 start + intervall
    plt.plot(times[start:end], errors[start:end])
    plt.xlabel('time [s]')
    plt.ylabel('error [rad]')
    plt.title('Error over time ' + name)
    plt.show()

# print(error_list)
```

```python
plot_list('')
plot_list('_kp=40')
plot_list('_kp=50')
plot_list('_kp=50_2')
plot_list('_kp=55')
plot_list('_kp=60')
```
